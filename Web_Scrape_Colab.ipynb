### Example 2

```python
import urllib.request
```

```python
response = urllib.request.urlopen('https://en.wikipedia.org/wiki/SpaceX')
html = response.read()
print(html)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
# Parsing the data
soup = BeautifulSoup(html, parser = 'html')
text = soup.get_text(strip = True)
print(text)
```

```python
tokens = [t for t in text.split()]
print(tokens)
```

```python
sr = stopwords.words('english')
clean_tokens = tokens[:]

for token in tokens:
  if token in stopwords.words('english'):
    clean_tokens.remove(token)
```

```python
clean_tokens
```

```python
## Check the Frequency of our Data

freq = nltk.FreqDist(clean_tokens)

# for key, value in freq.items():
#   print(str(key) + ' : ' + str(value))

freq.plot(20, cumulative = False)
```

```python
driver = gs.Chrome()
```

### Website to Scrape

```python
laptop_name = []
laptop_rating = []
laptop_processor = []
laptop_memory = []
laptop_os = []
laptop_storage = []
laptop_display = []
laptop_inlcusion = []
laptop_price = []
laptop_specs = []

driver.get('https://www.flipkart.com/search?q=gaming+laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_7_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_7_na_na_ps&as-pos=1&as-type=RECENT&suggestionId=gaming+laptop%7CLaptops&requestId=c1e0a154-c5dd-467f-989c-8f7b11ea8a2d&as-searchtext=gaming+&sort=price_desc&page=1')

laptops = driver.find_elements(By.CLASS_NAME, 'CGtC98')

for laptop in laptops:
  try:
    name = laptop.find_element(By.CLASS_NAME, 'KzDlHZ').text
    print(name)
    rating = laptop.find_element(By.CLASS_NAME, 'XQDdHH').text
    print(rating)
    price = laptop.find_element(By.CLASS_NAME, 'cN1yYO').text
    # print(price)

    ## Split the lines for the prices. Take the discounted price only
    lines_price = price.splitlines()
    array_line_price = []
    for line_price in lines_price:
      array_line_price.append(lines_price[0])

    specs = laptop.find_element(By.CLASS_NAME, 'G4BRas').text
    print(specs)

    ## Splitting the lines for the specs. Divide it on OS, Storage, Memory, Display, Processor
    lines = specs.splitlines()
    array_line = []
    for line in lines:
        array_line.append(line)
    if len(array_line) == 6:

        laptop_name.append(name) ## The name of the laptop
        laptop_price.append(array_line_price[0]) ## The price of the laptop after splitting the lines
        laptop_rating.append(rating)

        # Append per specs
        laptop_processor.append(array_line[0])
        laptop_memory.append(array_line[1])
        laptop_os.append(array_line[2])
        laptop_storage.append(array_line[3])
        laptop_display.append(array_line[4])
        laptop_inlcusion.append(array_line[5])

  except:
    pass

  sleep(3)
```

### EDA

```python
data = {'Name': laptop_name,'Price': laptop_price,'Rating':laptop_rating,'Processor':laptop_processor,'RAM':laptop_memory,
        'Operating System':laptop_os,'Storage':laptop_storage,'Display':laptop_display,'Inclusion':laptop_inlcusion}
```

```python
df = pd.DataFrame(data)

df.head()
```

```python
df.info()
```

```python
df['Price']=pd.to_numeric(df['Price'].replace({'\':'',',':''}, regex=True))
df['Rating'] = pd.to_numeric(df['Rating'])
```

```python
df.info()
```

```python
# Goal 1: Most Common Processor for Laptops
processor_count = df['Processor'].value_counts()
processor_count.head()
```

```python
# Goal 1: Most Common Processor for Laptops
processor_counts_top5 = processor_count.head()

plt.figure(figsize=(10, 5))
sns.barplot(x=processor_counts_top5.index[:5], y=processor_counts_top5.values)
plt.xticks(rotation=45, ha='right')
plt.xlabel("Processor Type")
plt.ylabel("Count")
plt.title("Most Common Processor for Laptops")
plt.show()
```

```python
# Price vs. Rating Correlation

plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["Price"], y=df["Rating"])
plt.xlabel("Price (in INR)")
plt.ylabel("Rating")
plt.title("Correlation Between Laptop Price and Rating")
plt.show()
```

```python
# Storage Distribution
storage_counts = df["Storage"].value_counts()

# Plot for Storage Distribution
plt.figure(figsize=(10, 5))
sns.barplot(x=storage_counts.index, y=storage_counts.values)
plt.xticks(rotation=45, ha='right')
plt.xlabel("Storage Type")
plt.ylabel("Count")
plt.title("Most Common Storage Type in Laptops")
plt.show()
```
